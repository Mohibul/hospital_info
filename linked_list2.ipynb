```python
import pandas as pd
```

```python
xls = pd.ExcelFile('D:/covid-19/hospital_info.xlsx')
df_names = pd.read_excel(xls,'names',header=None)
df_names.columns=['serial','names']
df_names
```

```python
df=pd.DataFrame()
for i in range(1,66):
    df_old=df
    xls = pd.ExcelFile('D:/covid-19/hospital_info.xlsx')
    df = pd.read_excel(xls,str(i))
    df=df.drop(['Unnamed: 0'],axis=1)

    df["Hospital's Name"]=df_names['names'].iloc[i-1]
    df=df.set_index("Hospital's Name")
    
    df = pd.concat([df_old,df])
    
from openpyxl import load_workbook
import os

filename = 'D:/covid-19/hospital_info.xlsx'
writer = pd.ExcelWriter(filename, engine='openpyxl')
if os.path.exists(filename):
    book = load_workbook(filename)
    writer.book = book

df.to_excel(writer,sheet_name='all-Hospitals')
writer.save()
writer.close()
```
